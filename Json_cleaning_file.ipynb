```python
def expand_data(row):
    list = row["data"]
    return {str(datetime.utcfromtimestamp(item['date'] / 1000).year): item['value'] for item in list}
```

```python
expanded_data = data.apply(lambda row: expand_data(row), axis=1)
```

```python
expanded_data
```

```python
expanded_df = pd.DataFrame(expanded_data.tolist())
```

```python
expanded_df
```

```python
def check_int(row):
    for item in row:
        if isinstance(item, int) or item == "(s)" or isinstance(item, float):
            pass
        else:
            return False
    return True


expanded_df.apply(lambda row: check_int(row), axis=1)
```

```python
def consolidate(row):
    list = []
    for item in row:
        if isinstance(item, int) or isinstance(item,float):
            list.append(item)

    if len(list) >= 1:
        return round(sum(list)/len(list),2)
    else:
        return 0

expanded_df.loc[:,"Estimated net-import"] = expanded_df.apply(lambda x: consolidate(x), axis=1)

            
```

```python
expanded_df
```

```python
df_final = pd.concat([data, expanded_df], axis=1)
```

```python
df_final['name'] = df_final['name'].str.extract('Electricity net imports, (.*), Annual')
df_final
```
