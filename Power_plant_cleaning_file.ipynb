```python
cleaned_data = data[(~pd.isna(data["estimated_generation_gwh"])) | (~pd.isna(data["generation_gwh_2013"]) | ~pd.isna(data["generation_gwh_2014"]) | ~pd.isna(data["generation_gwh_2015"]) | ~pd.isna(data["generation_gwh_2016"]) | ~pd.isna(data["generation_gwh_2017"]))]
```

```python
def consolidation(generation_gwh_2013, generation_gwh_2014, generation_gwh_2015, generation_gwh_2016, generation_gwh_2017, estimated_generation_gwh):
    array = np.array([generation_gwh_2013, generation_gwh_2014, generation_gwh_2015, generation_gwh_2016, generation_gwh_2017])
    array = array[~np.isnan(array)]
    if len(array) == 0:
        return estimated_generation_gwh
    else:
        
        return array.mean()

cleaned_data.loc[:, "cleaned_estimated_generation_gwh"]  = cleaned_data.apply(lambda x: 
                    consolidation(x["generation_gwh_2013"], x["generation_gwh_2014"], 
                                  x["generation_gwh_2015"], x["generation_gwh_2016"],
                                  x["generation_gwh_2017"], x["estimated_generation_gwh"]), axis=1)

    
```

```python
cleaned_data.drop(columns = ["generation_gwh_2013", "generation_gwh_2014", "generation_gwh_2015",
                             "generation_gwh_2016","generation_gwh_2017","estimated_generation_gwh"], inplace=True)
```
